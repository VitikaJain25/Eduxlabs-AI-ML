```python
import pandas as pd

df = pd.read_csv('/content/chipotle.tsv', sep = '\t')
df.head()
```

```python
df.dtypes
```

```python
#replace '$' with blank string '' in item_price column
# .str.replace()

df['item_price'] = df['item_price'].str.replace('$', '')
```

```python
df.head()
```

```python
# astype() - to convert the type of data in a column

df['item_price'] = df['item_price'].astype(float)
```

```python
df.head()
```

```python
df['item_price'].mean()
```

* Average, minimum and maximum order price of each item_name

```python
df.groupby('item_name')['item_price'].agg(['mean', 'min', 'max', 'count', 'sum']).sort_values('count', ascending = False)
```

* **Per Quantity price of each item**

```python
df.head()
```

```python
dfx = df.groupby('item_name')['quantity', 'item_price'].sum()
dfx.head()
```

```python
dfx['per_quantity_price'] = dfx.item_price / dfx.quantity
dfx.head()
```

```python
df.head()
```

```python
#.str.contains() - to find all instance where a particular character matches
df_chips = df[df['item_name'].str.contains('Chips')]
df_chips.head()
```

```python
# Filter all item with "Chicken" in its name.
df_chicken = df[df['item_name'].str.contains('Chicken')]
df_chicken.head()
```

**IMDB Movie Ratings**

```python
# Access data using URL
dfm = pd.read_csv('http://bit.ly/imdbratings')
dfm.head()
```

```python
# Filter the data based of following condition
# List of movies from 'Crime','Comedy' or 'Action' genre

dfm_cca = dfm[(dfm['genre'] == 'Crime') | (dfm['genre'] == 'Comedy') | (dfm['genre'] == 'Action')]
dfm_cca.head()
```

```python
dfm[dfm['genre'].isin(['Crime','Comedy','Action'])].head()   #isin() - to apply or condition on a single column
```

```python
# List of movies with star ratings more than 8 and duration less than 150

dfm_srd = dfm[(dfm['star_rating'] > 8) & (dfm['duration'] < 150)]
dfm_srd.head()
```

```python
# List of movies with genre 'Drama' or 'Action' with duration more than duration 150 and rating less than 7.5
dfm_gdr = dfm[((dfm['genre'] == 'Drama') | (dfm['genre'] == 'Action')) & (dfm['star_rating'] < 7.5) & (dfm['duration'] > 150)]
dfm_gdr.head()
```

```python
#List of movies in which 'Brad Pitt' was present as an actor

dfm_brad = dfm[dfm['actors_list'].str.contains('Brad Pitt')]
dfm_brad.head()
```
