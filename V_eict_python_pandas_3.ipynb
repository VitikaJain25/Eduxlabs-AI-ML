```python
import pandas as pd

df = pd.read_csv('/content/chipotle.tsv', sep = '\t')
df.head()
```

```python
df.dtypes
```

```python
#replace '$' with blank string '' in item_price column
# .str.replace()

df['item_price'] = df['item_price'].str.replace('$', '')
```

```python
df.head()
```

```python
# astype() - to convert the type of data in a column

df['item_price'] = df['item_price'].astype(float)
```

```python
df.head()
```

```python
df['item_price'].mean()
```

* Average, minimum and maximum order price of each item_name

```python
df.groupby('item_name')['item_price'].agg(['mean', 'min', 'max', 'count', 'sum']).sort_values('count', ascending = False)
```

* **Per Quantity price of each item**

```python
df.head()
```

```python
dfx = df.groupby('item_name')['quantity', 'item_price'].sum()
dfx.head()
```

```python
dfx['per_quantity_price'] = dfx.item_price / dfx.quantity
dfx.head()
```

```python
df.head()
```

```python
#Filter all items with "Chips" in its item name

df[df['item_name'] == "Chips"]
```

```python
#.str.contains() - to find all instance where a particular character matches
df_chips = df[df['item_name'].str.contains('Chips')]
df_chips.head()
```

```python
# Filter all item with "Chicken" in its name.
df_chicken = df[df['item_name'].str.contains('Chicken')]
df_chicken.head()
```

**Drinks By Country**

```python
dfd = pd.read_csv('/content/drinks.csv')
dfd.head()
```

```python
# How many countries are present
print(dfd['country'].unique())
print(len(dfd['country'].unique()))
```

```python
# How many continents are present
print(dfd['continent'].unique())
print(len(dfd['continent'].unique()))
```

```python
# Average, max, min consumption of beer, sprit and wine in each continent.

dfd.groupby('continent')['beer_servings', 'spirit_servings', 'wine_servings'].agg(['mean', 'min', 'max'])
```
