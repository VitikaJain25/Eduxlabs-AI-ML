**Missing Values in Data**

```python
df.isnull().sum()
```

**Drop the 'Cabin' column with 687 missing values which is close to 80% of total Data size.**

```python
# drop() => used to drop any column or row where axis defines row or column.

df1 = df.drop('Cabin', axis = 'columns')
df1.head()
```

**Drop the rows with missing values**

```python
# can be used to frop rows with null values, subset represents the feature we want to consider to identify

df2 = df1.dropna(subset = ['Embarked'])
df2.head()
```

```python
df2.isnull().sum()
```

```python
df3 = df2.copy()    #storing a copy of df2 in df3
```

**Fill the missing value of Age with Mean of age value**

```python
round(df2['Age'].mean(),2)
```

```python
# fillna() - to fill the null values or void spaces in any column

df2['Age'] = df2['Age'].fillna(round(df2['Age'].mean(),2))
```

```python
df2.isnull().sum()
```

```python
# Mean can sometimes mislead, So we use Median.

import numpy as np

n1 = np.array([1,2,2,3,4,5])
n2 = np.array([1,2,2,3,4,5,70])

print('n1-' , np.mean(n1), np.median(n1))
print('n2-' , np.mean(n2), np.median(n2))
```

**Analysis made using class of pasenger represents  class 1 passengers are relatively older than passengers travelling in class 2 and class 3.**

* So we can identify or check class of passenger in which he/she is travelling to fill the missing age value according to average age of that class.

```python
# Use filtering to seperate passengers with missing value of age and non missing values of age.
dfa0 = df3[df3['Age'].isnull()]
dfa1 = df3[~df3['Age'].isnull()] # Use ~ before conditon to get negation.
```

```python
dfa0.head()
```

```python
dfa1.head()
```

```python
def fix_missing_age_values(pclass):
  if pclass == 1:
    return 37
  elif pclass == 2:
    return 29
  else:
    return 24
```

```python
fix_missing_age_values(3)
```

```python
fix_missing_age_values(2)
```

```python
# apply() => To apply a user defined function on a Data column.

dfa0['Age'] = dfa0['Pclass'].apply(fix_missing_age_values)
```

```python
# Join back dfa0 and dfa1

df4 = pd.concat((dfa0, dfa1), axis = 'rows', ignore_index=True)
df4.head()
```

```python
df4.isnull().sum()
```

```python
# Create a python function which can accept the name and return surname

def get_surname(name):
  surname = name.split(',')[0]
  return surname
```
